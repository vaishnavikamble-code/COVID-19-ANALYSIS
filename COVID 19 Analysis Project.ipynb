```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset (Using a sample COVID-19 dataset from an online source)
url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv"
df = pd.read_csv(url)

# Step 2: Display basic information
print("Dataset Head:")
print(df.head())
print("\nDataset Info:")
print(df.info())

# Step 3: Filter data for a specific country (India)
india_df = df[df['Country'] == 'India']

# Step 4: Plot COVID-19 cases over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=india_df, x='Date', y='Confirmed', label='Confirmed Cases', color='blue')
sns.lineplot(data=india_df, x='Date', y='Recovered', label='Recovered Cases', color='green')
sns.lineplot(data=india_df, x='Date', y='Deaths', label='Death Cases', color='red')
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.title("COVID-19 Cases in India Over Time")
plt.legend()
plt.show()

# Step 5: Save the processed data (Optional)
india_df.to_csv("india_covid_data.csv", index=False)
```
